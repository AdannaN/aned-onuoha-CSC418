```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
img = cv.imread('img/download.jpg', 0)
rows, cols = img.shape
```

```python
#Plot the original image
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img)
M = np.float32([[1, 0, 100], [0, 1, 50]])
transl = cv.warpAffine(img, M, (cols, rows))
```

```python
#Plot the translated image
plt.subplot(1, 2, 2)
plt.title("Translated Image")
plt.imshow(transl)
```
